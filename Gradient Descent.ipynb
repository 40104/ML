## №1.1
Задание:Выбрать себе тестовую функцию из книжки Пантелеева и на ней все тестировать

#### Трехгорбая функция
#### f(x,y)= 2*x^2 - 1.05*x^4 + 1/6*x^6 +x*y +y^2
#### f(0,0) = 0
#### -5<=x,y<=5

```python
def f(x,y):
    return 2*x**2 - 1.05*(x**4) + 1/6*(x**6) +x*y +y**2

def dfx(x,y):
    return x**5-4.2*(x**3)+4*x+y

def dfy(x,y):
    return x+2*y

xx=random.uniform(-5, 5)
yy=random.uniform(-5, 5)

x_plt =np.arange(-5.0,5.0,0.1)
y_plt =np.arange(-5.0,5.0,0.1)

f_plt=np.array([[f(x,y) for x in x_plt]for y in y_plt])
```

#### Функция Функция Химмельблау
#### f(x,y)= (x^2+y-11)^2+(x+y^2-7)^2
#### f(3,2) = 0
#### f(-2.805118,3.131312) = 0

#### -5<=x,y<=5

```python
def f(x,y):
    return (x**2+y-11)**2+(x+y**2-7)**2

def dfx(x,y):
    return 4*x*(x*x+y-11)+2*x+2*y*y

def dfy(x,y):
    return 2*x*x+4*y*(x+y*y-7)+2*y-22

xx=random.uniform(-5, 5)
yy=random.uniform(-5, 5)

xx=5
yy=5

x_plt =np.arange(-5.0,5.0,0.1)
y_plt =np.arange(-5.0,5.0,0.1)

f_plt=np.array([[f(x,y) for x in x_plt]for y in y_plt])
```

## №1.2
Задание:самописный градиентный спуск на Python

```python
N=20
lmd1=0.01
lmd2=0.005
f_plt=np.array([[f(x,y) for x in x_plt]for y in y_plt])
print("x = {}, y = {}, F(x,y)= {}.".format(xx,yy, f(xx,yy)))
for i in range(N):
    xx=xx-lmd1*dfx(xx,yy)
    yy=yy-lmd2*dfy(xx,yy)
    print("x = {}, y = {}, F(x,y)= {}.".format(xx,yy, f(xx,yy)))
```

## №1.3
Задание:визуализация процесса спуска (matplolib, celluloid,plotly)

```python
%matplotlib notebook

N=20
lmd1=0.01
lmd2=0.005

plt.ion()
fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig)
ax.grid(True)
x,y=np.meshgrid(x_plt,y_plt)
ax.plot_surface(x,y,f_plt.T,cmap=cm.coolwarm,alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')

ax.scatter(xx,yy,f(xx,yy),c='blue')
for i in range(N):
    xx=xx-lmd1*dfx(xx,yy)
    yy=yy-lmd2*dfy(xx,yy)
    ax.scatter(xx,yy,f(xx,yy),c='red')
    
    fig.canvas.draw()
    fig.canvas.flush_events()
    

plt.ioff()

ax.scatter(xx,yy,f(xx,yy),c='black')
print("x = {}, y = {}, F(x,y)= {}.".format(xx,yy, f(xx,yy)))

plt.show()

try:
    while fig.number in plt.get_fignums():
        plt.pause(0.2)
except:
    plt.close(fig.number)
    raise
```

## №1.6
Задание:Реализовать одну адаптивную модификацию ГС

```python
%matplotlib notebook

N=50
lmd1=0.1
lmd2=0.05
Gx=0
Gy=0

plt.ion()
fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig)
ax.grid(True,auto_add_to_figure=False)
x,y=np.meshgrid(x_plt,y_plt)
ax.plot_surface(x,y,f_plt.T,cmap=cm.coolwarm,alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')

ax.scatter(xx,yy,f(xx,yy),c='blue')
for i in range(N):
    Gx+=dfx(xx,yy)**2
    Gy+=dfy(xx,yy)**2
    xx=xx-dfx(xx,yy)*lmd1/(Gx+0.000000001)**0.5
    yy=yy-dfy(xx,yy)*lmd2/(Gy+0.000000001)**0.5
    ax.scatter(xx,yy,f(xx,yy),c='red')
    
    fig.canvas.draw()
    fig.canvas.flush_events()
    

plt.ioff()

ax.scatter(xx,yy,f(xx,yy),c='black')
print("x = {}, y = {}, F(x,y)= {}.".format(xx,yy, f(xx,yy)))

plt.show()

try:
    while fig.number in plt.get_fignums():
        plt.pause(0.2)
except:
    plt.close(fig.number)
    raise
```

## №1.7
Задание:Реализовать один трюк для ГС (метод старта, динамику изменения скорости обучения)

```python
#Динамичная скорость обучения
%matplotlib notebook

N=20
lmd1=0.01
lmd2=0.005
e=0.1
plt.ion()
fig = plt.figure(figsize=(8, 8))
ax = Axes3D(fig)
ax.grid(True)
x,y=np.meshgrid(x_plt,y_plt)
ax.plot_surface(x,y,f_plt.T,cmap=cm.coolwarm,alpha=0.5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')

ax.scatter(xx,yy,f(xx,yy),c='blue')
for i in range(N):
    try:
        xx=xx-lmd1*dfx(xx,yy)
    except ZeroDivisionError:
        lmd1*=e
        xx=xx-lmd1*dfx(xx,yy)
    try:
        yy=yy-lmd2*dfy(xx,yy)
    except ZeroDivisionError:
        lmd2*=e
        yy=yy-lmd2*dfy(xx,yy)
    ax.scatter(xx,yy,f(xx,yy),c='red')
    
    fig.canvas.draw()
    fig.canvas.flush_events()
    

plt.ioff()

ax.scatter(xx,yy,f(xx,yy),c='black')
print("x = {}, y = {}, F(x,y)= {}.".format(xx,yy, f(xx,yy)))

plt.show()

try:
    while fig.number in plt.get_fignums():
        plt.pause(0.2)
except:
    plt.close(fig.number)
    raise
```
